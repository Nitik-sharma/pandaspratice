Q4: Salary Cleaning
Q4.1 â€” Extract minimum salary into a new column named MinSalary.
Q4.2 â€” Extract maximum salary into a new column MaxSalary.
Q4.3 â€” Convert both into proper integers (44000, 99000, etc.).


```python


#4.1
data["MinSalary"]=(
    data['Salary']
    .str.replace("$","",regex=False)
    .str.replace("k","",regex=False)
    .str.split("-",expand=True)[0]
    .astype(int)*1000

)

#4.2

data["MaxSalary"]=(
    data['Salary']
    .str.replace("$","",regex=False)
    .str.replace("k","",regex=False)
    .str.split("-",expand=True)[1]
    .astype(int)*1000
)
data.head()



```

Now Let's Move to LEVEL 2 â€“ Q5 (Location Cleaning)


```python
data['Location']=data['Location'].str.rsplit(" ",n=1).str.join(",")
data[["City","Country"]]=data["Location"].str.rsplit(",",n=1,expand=True)


data.head()
```

```python
data['City']=data['City'].str.replace(","," ",1)
data.head()
```

Next Task (LEVEL 2 â€“ Q6) 
Q6 â€” Replace:

Rating: -1 â†’ NaN

Established: -1 â†’ NaN

Easy Apply: -1 â†’ "UNKNOWN"

```python
data["Rating"]=data["Rating"].replace(-1,np.nan)
data["Established"]=data["Established"].replace(-1,np.nan)
data['Easy Apply']=data["Easy Apply"].str.replace("NaN","UNKNOWN")
data.head()
```

NEXT LEVEL (LEVEL 3 â€“ Advanced Pandas)

Now answer Q7:

Q7.1 â€” Find average Age for each Country.
Q7.2 â€” Find minimum and maximum salary (MinSalary, MaxSalary) by Country.

```python
average=data.groupby("Country")["Age"].mean().round(2)
print("Average of Salary  each country : ",average)
maxSalary=data.groupby("Country")["MaxSalary"].max()

minSalary=data.groupby("Country")["MinSalary"].min()
print("Max Salary= ",maxSalary)
print("Min Salary= ",minSalary)
```

Now Let's Move to LEVEL 3 â€“ Q8 (Outlier Detection using IQR)

Write code for:

Q8.1 â€” Compute IQR for Rating
Q8.2 â€” Get all rows where Rating is an outlier

```python
Q1=data["Rating"].quantile(0.25)
Q3=data["Rating"].quantile(0.75)

IQR=Q3-Q1

lower=Q1-1.5*IQR
upper=Q3+1.5*IQR


outliner=data[(data["Rating"]<lower) | (data["Rating"]>upper)]
print(outliner)
```
