Q4: Salary Cleaning
Q4.1 — Extract minimum salary into a new column named MinSalary.
Q4.2 — Extract maximum salary into a new column MaxSalary.
Q4.3 — Convert both into proper integers (44000, 99000, etc.).


```python


#4.1
data["MinSalary"]=(
    data['Salary']
    .str.replace("$","",regex=False)
    .str.replace("k","",regex=False)
    .str.split("-",expand=True)[0]
    .astype(int)*1000

)

#4.2

data["MaxSalary"]=(
    data['Salary']
    .str.replace("$","",regex=False)
    .str.replace("k","",regex=False)
    .str.split("-",expand=True)[1]
    .astype(int)*1000
)
data.head()



```

Now Let's Move to LEVEL 2 – Q5 (Location Cleaning)


```python
data["Location"] = data["Location"].str.replace(" ", ",", 1)

data[["City","Country"]]=data['Location'].str.split(",",expand=True)


data.head()
```
